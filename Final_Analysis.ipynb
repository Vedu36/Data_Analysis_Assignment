```python
import pandas as pd
```

```python
# Load the Excel file
xls = pd.ExcelFile("Assignment.xlsx")

user_details = pd.read_excel(xls, 'UserDetails.csv')
cooking_sessions = pd.read_excel(xls, 'CookingSessions.csv')
order_details = pd.read_excel(xls, 'OrderDetails.csv')

print(user_details.head())
print(cooking_sessions.head())
print(order_details.head())
```

# Data transforming and cleaning

```python
# Filling nan values with mean value
average_rating = order_details['Rating'].mean().round(1)

order_details['Rating'].fillna(average_rating, inplace=True)
```

```python
print(order_details)
print("Sum of null values",order_details['Rating'].isnull().sum())
```

```python
# Merge cooking_sessions with order_details on 'Session ID'
merged_sessions_orders = pd.merge(cooking_sessions, order_details, on='Session ID', how='left')

merged_sessions_orders.head()
```

```python
# Renaming the columns
merged_sessions_orders.rename(columns={'User ID_x': 'User ID',"Dish Name_x":"Dish Name","Meal Type_x":"Meal Type",}, inplace=True)

# Merge merged_sessions_orders with user_details on 'Session ID'
final_data = pd.merge(merged_sessions_orders, user_details, on='User ID', how='left')

final_data.columns
```

```python
# Droping the Dublicate and unwanted columns
final_data.drop(['Meal Type_y', 'Dish Name_y','User ID_y','Phone', 'Email'], axis=1, inplace=True)

final_data
```

# Starting Analysis

Relation: Average Duration and Average Rating for Different Dishes

```python
# Calculate the average Duration for each dish
average_duration_per_dish = final_data.groupby('Dish Name')['Duration (mins)'].mean().reset_index()

average_duration_per_dish.rename(columns={'Duration (mins)': 'Average Duration (mins)'}, inplace=True)
print(average_duration_per_dish)
```

```python
# Calculate the average rating for each dish
average_rating_per_dish = final_data.groupby('Dish Name')['Session Rating'].mean().reset_index()

average_rating_per_dish.rename(columns={'Session Rating': 'Average Rating'}, inplace=True)
print(average_rating_per_dish)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot Average Duration on the first y-axis
color = 'tab:blue'
ax1.set_xlabel('Dish Name', fontsize=14)
ax1.set_ylabel('Average Duration (mins)', fontsize=14, color=color)
bars = ax1.bar(average_duration_per_dish['Dish Name'], average_duration_per_dish['Average Duration (mins)'], color=color, alpha=0.6, label='Average Duration (mins)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(average_duration_per_dish['Dish Name'], rotation=45, ha='right')

# Add text labels for the average duration values
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom', color='black')

# Create a second y-axis to plot Average Rating
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Average Rating', fontsize=14, color=color)
ax2.plot(average_rating_per_dish['Dish Name'], average_rating_per_dish['Average Rating'], color=color, marker='o', label='Average Rating')
ax2.tick_params(axis='y', labelcolor=color)

# Add titles and labels
plt.title('Average Duration and Average Rating for Different Dishes', fontsize=16)

# Add legends
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

Most Common Dishes for Each Meal Type

```python
# Calculate the count of each dish within each meal type
meal_dish_count = final_data.groupby(['Meal Type', 'Dish Name']).size().reset_index(name='Count')

# Find the most common dishes for each meal type
most_common_dishes = meal_dish_count.loc[meal_dish_count.groupby('Meal Type')['Count'].apply(lambda x: x == x.max())]
print(most_common_dishes)
```

```python
# Set the plot style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Meal Type', y='Count', hue='Dish Name', data=most_common_dishes, dodge=True, palette='viridis')
plt.title('Most Common Dishes for Each Meal Type', fontsize=16)
plt.xlabel('Meal Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Dish Name')
plt.show()
```

Average Amount Spent on Each Dish

```python
# Calculate the average amount spent on each dish
average_amount_per_dish = final_data.groupby('Dish Name')['Amount (USD)'].mean().reset_index()
average_amount_per_dish.rename(columns={ 'Amount (USD)': 'Average Amount (USD)'}, inplace=True)

# Sorting it
average_amount_per_dish_sorted = average_amount_per_dish.sort_values(by='Average Amount (USD)')
print(average_amount_per_dish_sorted)
```

```python
# Set the plot style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Average Amount (USD)', y='Dish Name', data=average_amount_per_dish_sorted, palette='coolwarm')
plt.title('Average Amount Spent on Each Dish', fontsize=16)
plt.xlabel('Average Amount (USD)', fontsize=14)
plt.ylabel('Dish Name', fontsize=14)
plt.show()
```
